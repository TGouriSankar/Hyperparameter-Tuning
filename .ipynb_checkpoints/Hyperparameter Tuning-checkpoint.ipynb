### Model

```python
x = wine.drop(['quality'],axis=1)
y = wine['quality']
print("The shape of x :{} The shape of y :{}".format(x.shape,y.shape))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print("The shape of x_trtain :{} The shape of y_train :{}".format(x_train.shape,y_train.shape))
print("The shape of x_test :{} The shape of y_test :{}".format(x_test.shape,y_test.shape))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
y_ped=model.predict(x_test)
y_ped
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_ped,y_test)
```

### GridSearchCV

```python
grid_param = {
    "n_estimators" : [90,100],
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,10,1),
    'min_samples_leaf' : range(1,5,1),
    'min_samples_split': range(2,5,1),
    'max_features' : ['auto','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator=model,param_grid=grid_param,cv=5)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
model_grid = RandomForestClassifier(criterion= 'gini',max_depth= 9,max_features= 'auto',min_samples_leaf= 1,min_samples_split= 3,n_estimators= 90,random_state=1000)
model_grid.fit(x_train,y_train)
```

```python
y_pred = model_grid.predict(x_test)
print(y_pred)
```

```python
accuracy_score(y_pred,y_test)
```

### model outliers capping

```python
wine.columns
```

```python
from feature_engine.outliers import Winsorizer
```

```python
fix = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='fixed acidity')
wine['fixed acidity'] = fix.fit_transform(wine[['fixed acidity']])
```

```python
sns.boxplot(wine['fixed acidity'])
```

```python
vol = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='volatile acidity')
wine['volatile acidity'] = vol.fit_transform(wine[['volatile acidity']])
sugar = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='residual sugar')
wine['residual sugar'] = sugar.fit_transform(wine[['residual sugar']])
cho = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='chlorides')
wine['chlorides'] = cho.fit_transform(wine[['chlorides']])
free = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='free sulfur dioxide')
wine['free sulfur dioxide'] = free.fit_transform(wine[['free sulfur dioxide']])
total = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='total sulfur dioxide')
wine['total sulfur dioxide'] = total.fit_transform(wine[['total sulfur dioxide']])
den = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='density')
wine['density'] = den.fit_transform(wine[['density']])
ph = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='pH')
wine['pH'] = ph.fit_transform(wine[['pH']])
sulp = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='sulphates')
wine['sulphates'] = sulp.fit_transform(wine[['sulphates']])
hol = Winsorizer(capping_method='iqr',tail='both',fold=1.5,variables='alcohol')
wine['alcohol'] = hol.fit_transform(wine[['alcohol']])
```

```python
for i in wine.columns:
    sns.boxplot(wine[i],color="red")
    plt.show()
```

```python
x1 = wine.drop(['quality'],axis=1)
y1 = wine['quality']
print("The shape of x1 :{} The shape of y1 :{}".format(x1.shape,y1.shape))
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size=0.2,random_state=42)
print("The shape of x1_trtain :{} The shape of y1_train :{}".format(x1_train.shape,y1_train.shape))
print("The shape of x1_test :{} The shape of y1_test :{}".format(x1_test.shape,y1_test.shape))
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=100,random_state=42)
model1.fit(x_train,y_train)
```

```python
y1_pred=model1.predict(x1_test)
y1_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y1_pred,y1_test)
```
